# Perceptron Learning Algorithm for different sized datasets

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
directory = 'D:\Courses\Fall 19\ELEG 815 Statistical Learning\HW4\Homework4Files'
files = ['\inputData.csv','\inputData20.csv','\inputData100.csv','\inputData1000.csv']
#files = ['\inputData.csv']
```

```python
def Target_Function(x_1,x_2):
    return int(np.sign(5*x_1 + 2*x_2 + 1))

def Evaluate_on_Target(inputData,true_labels):
    for i in range(len(inputData)):
        x_1 = inputData.iloc[i][0]
        x_2 = inputData.iloc[i][1]
        true_labels[i] = Target_Function(x_1,x_2)
    return true_labels

def Set_Colors_Array(true_labels):
    col = [None]*len(true_labels)
    for i in range(len(col)):
        if true_labels[i] == -1:
            col[i] = 'r'
        else:
            col[i] = 'g'
    return col

def Scatter_Plot(inputData,file):
    x_1 = inputData[0]
    x_2 = inputData[1]
    y = (-1-5*x_1)/2
    plt.plot(x_1,y)                   # Plot the true function (1+5*x1+2*x2)
    plt.scatter(x_1,x_2,c=colors)     # Plot the input data points
    plt.legend(['Target'])
    plt.xlabel(file[1:])
    plt.show()

def Intialize_Data_Structures(inputData,true_labels):
    inputArray = np.array(inputData)
    np.random.seed(1111)
    bias = 0.1
    W_rand = np.random.normal(0,0.0005,(np.size(inputArray,1) +1,))        # Get number of columns in inputArray to get dimensionality of data
    W_rand[0] = bias
    W = W_rand
    X = np.zeros((len(true_labels),3))
    X[:,1:] = inputArray
    X[:,0] = 1
    return W,X
    
def isMisclassified(point,true_label,W):
    if np.sign(np.dot(W.T,point)) != true_label:
        return True
    else:
        return False

def Perceptron_Learner(W,X,true_labels,no_iterations):
    for j in range(no_iterations):
        for i,point in enumerate(X):
            if isMisclassified(point,true_labels[i],W):
                yx =  np.dot(true_labels[i],X[i])        
                W += yx
    return W

def Accuracy(W,X,true_labels):
    cnt = 0
    for i,point in enumerate(X):
        if np.sign(np.dot(W.T,point)) != true_labels[i]:
            cnt += 1
    acc = (1 - cnt/len(true_labels))*100
    return acc

def Plot_Predictions(inputData,file):
    x_1 = inputData[0]
    x_2 = inputData[1]
    y = (-1-5*x_1)/2
    z = (-W[0]-W[1]*x_1)/W[2]
    plt.plot(x_1,y)
    plt.plot(x_1,z)
    plt.scatter(x_1,x_2,c = colors)
    plt.legend(['f(x)','g(x)'])
    plt.xlabel(file[1:])
    plt.show()
```

```python
target_vector = np.array([1,5,2])
for key,file in enumerate(files):
    plt.figure(key)
    inputData = pd.read_csv(directory+file,header=None)
    true_labels = [0]*len(inputData)
    true_labels = Evaluate_on_Target(inputData,true_labels)
    colors = Set_Colors_Array(true_labels)
    Scatter_Plot(inputData,file)
    acc = 0
    iteration = 0
    while acc != 100.0:
        iteration += 1
        W,X = Intialize_Data_Structures(inputData,true_labels)   # Weights are intialized again and again in each iteration until acc = 100 %
        Perceptron_Learner(W,X,true_labels,iteration)      
        acc = Accuracy(W,X,true_labels)
    Plot_Predictions(inputData,file)
    print('Final Weight Vector :',W)
    print('Norm of final Weight vector :',W/np.linalg.norm(W))
    print('Norm of Target vector :',target_vector/np.linalg.norm(target_vector))
    print('Iterations required for convergence :',iteration)
```
