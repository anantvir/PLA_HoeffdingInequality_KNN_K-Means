```python
import numpy as np
from scipy import spatial
from PIL import Image
import random
import operator
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
path = 'D:\\Courses\\Fall 19\\ELEG 815 Statistical Learning\\HW4\\Homework4Files\\'
img = Image.open(path + 'Image.png','r')
img = img.resize((400,400))
img = img.convert('RGB')
width, height = img.size
pix_lst = list(img.getdata())
k_list = [3,10,20]
```

```python
for k in k_list:
    print('Compressing Image into k = {} clusters'.format(k))
    # ------------------ Create list of Clusters, Example [[],[],[]] for k = 3-------------------------
    clusters = [[] for i in range(k)]
    inital_centers = [0]*k
    for i in range(k):   
        flag = True
        while flag:
            ind = random.randint(1,len(pix_lst))
            if pix_lst[ind] not in inital_centers:
                inital_centers[i] = pix_lst[ind]
                flag = False
            else:
                continue
    print('Initial Centers :',inital_centers)
    old_Centers = [0,0,0]
    new_centers = inital_centers
    iteration = 1
    while old_Centers != new_centers:
        print('Converging Iteration :',iteration)
        iteration += 1
        old_Centers = new_centers
        for pixel in range(len(pix_lst)):
            distances = []
            for center in inital_centers:
                dist = spatial.distance.euclidean(center,pix_lst[pixel])
                distances.append(dist)
            ind = distances.index(min(distances))
            clusters[ind].append(pix_lst[pixel]) 
        #------------------Calculate New Centers--------------------------
        # https://stackoverflow.com/questions/497885/python-element-wise-tuple-operations-like-sum
        new_centers = []
        for cluster in clusters:
            t = (0,0,0)
            for pixelTuple in cluster:
                t = tuple(map(operator.add,t,pixelTuple))
            new_center = tuple(map(lambda x: int(x/len(cluster)),t))
            new_centers.append(new_center)

    """Reconstruct Image from Pixel Tuples"""
    newImage = Image.new('RGB',(width,height))
    rimg = newImage.load()
    print('Reconstructing Image for k={}'.format(k))
    for i in range(height):
        for j in range(width):
            for index,cluster in enumerate(clusters):
                if(pix_lst[i*width+j]) in cluster:
                    rimg[j,i] = new_centers[index]
    newImage.save(path+"newImage_k{}.png".format(k), "PNG")

print('Finished !')
```

```python
img=mpimg.imread(path+'newImage_k3.png')
imgplot = plt.imshow(img)
plt.xlabel('K = 3')
```

# Results and Conclusions

As value of k increases, we see that image becomes more and more detailed. This is because we can now assign pixels to more clusters, which means more colours. Depending on the image, we can choose a value of k that can accurately represent the image. E.g if we choose an image having 2 dominant colors, then k = 2 might suffice to represent the image(means the image can be seen and understood by everyone without any loss of information)
